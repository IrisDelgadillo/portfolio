```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data
students = pd.read_csv('students.csv')
```

```python

# Print first few rows of data
print(students.head())
# Print summary statistics for all columns
print(students.describe())
# Calculate mean
mean = students.math_grade.mean()
# Calculate median
median = students.math_grade.median()
# Calculate mode
mode = students.math_grade.mode()
# Calculate range
ranges = students.math_grade.max() - students.math_grade.min()

print(mode)
print(ranges)
```

```python

# Calculate standard deviation
standard = students.math_grade.std()
print(standard)
```

```python

# Create a histogram of math grades
sns.histplot(x='math_grade', data = students)
plt.show()
plt.clf()
```

```python

# Create a box plot of math grades
sns.boxplot(x = 'math_grade', data = students)
plt.show()
plt.clf()
```

```python


# Calculate number of students with mothers in each job category
mjobs = students.Mjob.value_counts()
print(mjobs)
# Calculate proportion of students with mothers in each job category
mjobs_prop = students.Mjob.value_counts(normalize = True)
print(mjobs_prop)
```

```python

# Create bar chart of Mjob
sns.countplot(x = 'Mjob', data = students)
plt.show()
plt.close()
```

```python
# Create pie chart of Mjob
students.Mjob.value_counts().plot.pie()
plt.show()
plt.close()
```
